## Температура

```python
q1t = main_factors.rectal_temperature.quantile(0.25)
q3t = main_factors.rectal_temperature.quantile(0.75)
iqrt = q3t - q1t
lower_bound_t = q1t - (1.5 * iqrt) 
upper_bound_t = q3t + (1.5 * iqrt)

remove_outliers_t = main_factors[main_factors.rectal_temperature.between(lower_bound_t, upper_bound_t, inclusive=True)].sort_values('rectal_temperature')

print(lower_bound_t, upper_bound_t)
```

```python
pd.concat([main_factors, remove_outliers_t]).drop_duplicates(keep=False).dropna()
```

```python
# Предполагаем, что уровень боли зависит от пульса, что частично подтверждается данными ниже
# Будем заполнять по интервалам, предположим что пульс 30-55 относится к первой группе боли, 55-65 к 3, 65-80 к 2, 80-85 к 4 и 85 и выше к 5. 
# Также мы значем, что для всех незаполненных значений боли у нас есть значения пульса, что приведет к полному заполнению значений боли

print(main_factors_nn.groupby('pain').pulse.mean())
main_factors_nn[main_factors_nn[['pain']].isnull().any(1)]
```

```python
# Заполняем значения для боли

main_factors_nn.pain = np.where((main_factors_nn.pain.isnull) & (55 > main_factors_nn.pulse > 30), 1, main_factors_nn.pain)


```

```python
def pain(pa, pu):
    if pa.isnull & (55 > pu >= 30):
        return 1
    elif pa.isnull & (80 > pu >= 65):
        return 2
    elif pa.isnull & (65 > pu >= 55):
        return 3
    elif pa.isnull & (85 > pu >= 80):
        return 4
    else:
        return 5
    
main_factors_nn.pain = main_factors_nn.apply(pain(main_factors_nn.pain, main_factors_nn.pulse), axis = 1)
        
    
```
